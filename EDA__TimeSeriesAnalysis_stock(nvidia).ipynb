### **Time Rsampling**

```python
data_nvdia.resample(rule='A').min()
```

```python
## A = year end frequency
#https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78

data_nvdia.resample(rule='A').max()
```

```python
data_nvdia.resample(rule='A').max()['High'].plot()
```

```python
type(data_nvdia.resample(rule='A').max())
```

```python
data_nvdia['Open'].resample(rule='BA').mean().plot(kind = 'bar')
```

### rolling and expanding

```python
#rolling , expanding
data_nvdia['High'].rolling(11).mean().head(12)
```

```python
data_nvdia['Open: 30 days rolling']= data_nvdia['Open'].rolling(30).mean()
```

```python
data_nvdia
```

```python
data_nvdia[['Open','Open: 30 days rolling']].plot(figsize=(12,6)) #smotthens the curve
```
