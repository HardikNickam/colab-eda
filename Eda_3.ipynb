```python
## feature engineerng process

# take day month and year from date
final_df['Day']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Day']=final_df['Day'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head()
```

```python
# final_df['Arrival_Time'].str.split(' ').str[0]
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
final_df['Arrival_Time']
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[0])
final_df['Arrival_min']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[1])
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df.head()
final_df.info()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head()
```

```python
final_df['Dep_Time_hour']=final_df['Dep_Time'].apply(lambda x:x.split(':')[0])
final_df['Dep_Time_min']=final_df['Dep_Time'].apply(lambda x:x.split(':')[1])
final_df['Dep_Time_hour']=final_df['Dep_Time_hour'].astype(int)
final_df['Dep_Time_min']=final_df['Dep_Time_min'].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.drop('Route',axis=1,inplace=True)
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['duration_hour'] = final_df['Duration'].str.split('h').str[0]
final_df.head()
```

```python
final_df['duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0)
final_df.head()
```

```python
final_df.drop(6474,inplace=True,axis=0)
final_df.drop(2660,inplace=True,axis=0)
```

```python
final_df['duration_hour']=final_df['duration_hour'].astype(int)
final_df['duration_min']=final_df['duration_min'].astype(int)
final_df['duration_hour']=final_df['duration_hour']*60
final_df['total_duration']= final_df['duration_hour'] + final_df['duration_min']
final_df.drop('Duration',inplace=True,axis=1)
final_df.head()
```
