```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df= pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
 final_df.dtypes
```

```python
final_df.columns
```

```python
CountryNames=final_df.Country.value_counts().index
```

```python
Countryvalue=final_df.Country.value_counts().values
```

```python
plt.pie(Countryvalue[:3],labels=CountryNames[:3],autopct='%1.5f%%')  #top 3 values n country using zomato
```

observation:zomato masximum trnsactions are from India After that the Usa and Then the UNited Kingdom

```python
final_df.columns
```

```python
ratings= final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating counts"})
```

```python
ratings
```

conclusiosn:

Whenever rating is ntween 4.5 - 4.9 --->food was excellent

when ratings from 4.0-4.4--->food was very good

when ratings from 3.4-3.9--->food was good

when ratings from 2.5-2.9--->food was average

when ratings from 1.8-2.4--->food was poor

```python
ratings
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize'] = (12,6)



sns.barplot(x="Aggregate rating", y="Rating counts",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating counts",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
